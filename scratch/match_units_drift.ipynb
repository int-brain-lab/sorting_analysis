```python
# make violin plots
#matcher.load_matches('/match_to_sort/')
#matcher.load_matches('/match_to_truth/')
print ("n_units: ", len(matcher.n_spikes))

def get_violin_data(matcher, 
                    selected_unit,
                    root_dir):
    
    n_matches2 = 1
    
    #spike_train_gt = np.load(root_dir + 'ground_truth/spike_train_ground_truth.npy')
    #templates_sorted = np.load(root_dir + 'templates_reloaded_good.npy')
    spike_train_sorted = matcher.spike_train_sorted
    templates_sorted = matcher.templates_sorted
    spike_train_gt = matcher.spike_train_gt
    templates_gt = matcher.templates_gt
        
    max_chans_sorted = templates_sorted.ptp(1).argmax(1)

    # fix n-chans and sample rate for computations below
    n_chans = templates_sorted.shape[2]
    sample_rate = 30000
    
    # load units
    #matcher = Match_to_ground_truth(root_dir, spike_train_sorted, templates_sorted)

    matched_units = np.array(matcher.ids_matched[selected_unit][1])
    #print ("matched_units: ", matched_units)

    # count # of spikes in each matching unit;
    n_spk = []
    for k in range(len(matcher.matched_spikes_array[selected_unit])):
        n_spk.append(len(matcher.matched_spikes_array[selected_unit][k]))
    
    # select top 5 matching units by size and plot them;
    n_spk = np.hstack(n_spk)
    idx_sorted = np.argsort(n_spk)[::-1][:n_matches2]

    # ****************** GROUND TRUTH UNIT COMPUTATION ***************
    # load spikes for ground truth unit:
    idx = np.where(spike_train_gt[:,1]==selected_unit)[0]
    spk_gt = spike_train_gt[idx,0]
    tot_spikes_gt = spk_gt.shape[0]
    
    # count # of spikes in each matching unit;
    n_spk = []
    for p in range(len(matcher.matched_spikes_array[selected_unit])):
        n_spk.append(len(matcher.matched_spikes_array[selected_unit][p]))

    # select top match unit by size <-- MIGHT NEED TWEAKING
    # SHOULD SELECT TOP MATCH BY PURITY or COMPLETENESS
    n_matches2 = 1
    n_spk = np.hstack(n_spk)
    idx_sorted = np.argsort(n_spk)[::-1][:n_matches2]

    # ************ SCATTER PLOT BEST X MATCHES **************
    # track TP rates; 
    TP_rate = []
    # look only at the best match:
    #for k in range(len(ptps_sorted)):
    #for k in range(len(idx_sorted)):
    k=0   
    size = np.sort(matcher.venn_array[selected_unit])[::-1][k]
    tot_spikes = matcher.n_spikes_array[selected_unit]  # spikes in the ground truth unit
    tp_rate = size/float(tot_spikes)   

    id_ = idx_sorted[k]
    print ("Selected_unit: ", selected_unit, " , matching id: ", id_)

    # compute purity and compleness
    n_spikes_sorted_unit = np.where(matcher.spike_train_sorted[:,1]==
                      matched_units[id_])[0].shape[0]

    purity = size/float(n_spikes_sorted_unit)*100.
    completeness = size/float(tot_spikes)*100.


    TP_rate = tp_rate
                
    #for k in range(n_matches2):
    k=0
    idx_gt = np.where(matcher.spike_train_gt[:,1]==selected_unit)[0]
    times_gt = matcher.spike_train_gt[idx_gt,0]

    idx_sort = np.where(matcher.spike_train_sorted[:,1]==matched_units[idx_sorted[k]])[0]
    times_sort = matcher.spike_train_sorted[idx_sort,0]

    vintersect = n_spk[idx_sorted][k]
    v1 = times_gt.shape[0] - vintersect
    v2 = times_sort.shape[0] - vintersect

    FP_rate = v2/float(times_sort.shape[0])
        
    return TP_rate, FP_rate, purity, completeness, id_

# reload original spike trains
print ("RELOADING DATA AGAINST TRUTH")
matcher.load_data(spike_train_sorted, templates_sorted)
matcher.load_matches('/match_to_truth/')

TP=[]
FP=[]
purity = []
completeness = []
matching_ids = []
for selected_unit in range(len(matcher.n_spikes)):

    TP_, FP_, purity_, completeness_, id_ = get_violin_data(matcher,
                                                       selected_unit,
                                                       root_dir)

    TP.append(TP_)
    FP.append(FP_)
    purity.append(purity_)
    completeness.append(completeness_)
    matching_ids.append(id_)

# reload matches for all units
print ("RELOADING DATA AGAINST SORT")
matcher.load_data(spike_train_sorted, templates_sorted)
matcher.load_matches('/match_to_sort/')

spike_train_sorted = matcher.spike_train_gt.copy()
templates_sorted = matcher.templates_gt.copy()
spike_train_gt = matcher.spike_train_sorted.copy()
templates_gt = matcher.templates_sorted.copy()

matcher.spike_train_gt = spike_train_gt
matcher.templates_gt = templates_gt
matcher.spike_train_sorted = spike_train_sorted
matcher.templates_sorted = templates_sorted
   
TP_all=[]
FP_all=[]
purity_all = []
completeness_all = []
matching_ids_all = []
for selected_unit in range(len(matcher.n_spikes)):
    
    TP_, FP_, purity_, completeness_, id_ = get_violin_data(matcher,
                                                       selected_unit,
                                                       root_dir)

    TP_all.append(TP_)
    FP_all.append(FP_)
    purity_all.append(purity_)
    completeness_all.append(completeness_)
    matching_ids_all.append(id_)
```
